```python
from flask import Flask ,  url_for
```

```python
app= Flask(__name__)
```

```python
@app.route('/')
def index():
    return 'Welcome to the index page!'

@app.route('/user/<username>')
def user(username):
    return f'Hello, {username}!'

if __name__ == '__main__':
    app.run(host = "0.0.0.0")
    
index_url = url_for('index')  # Output: /
user_url = url_for('user', username='john')  # Output: /user/john
```
